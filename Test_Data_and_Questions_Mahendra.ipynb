## 4) Create a class and methods to add, remove and update Transactions & compute net profit of a transaction for transaction_details model above.

```python
# solution for question 4:
class Transaction:
    def __init__(self, txn_id, branch, product_sku, product_cost, cust_id, sold_amt):
        self.transaction_id = txn_id
        self.branch = branch
        self.product_sku = product_sku
        self.product_cost = product_cost
        self.customer_id = cust_id
        self.sold_amount = sold_amt

    def get_net_profit(self):
        return self.sold_amount - self.product_cost



class TransactionDetails:
    def __init__(self):
        self.details = []
    def add(self, txn):
        self.details[txn.transaction_id] = txn
    def remove(self, txn_id):
        self.details.pop(txn_id)
    def update(self, txn_id, new_txn):
        self.details[txn_id]= new_txn
```

## 5) Write the function which produces following output
[1, 9, 36, 84, 126, 126, 84, 36, 9, 1]

```python
# solution for question 5:
def gen(n,r=[]):
    for x in range(n):
        l = len(r)
        r = [1 if i == 0 or i == l else r[i-1]+r[i] for i in range(l+1)]
        yield r

l = list(gen(10))
print(l[-1])
```
